2. Obtener los 3 clientes que han realizando el mayor gasto total con el monto.

```python
clientes_mas_gasto = sql.conexion_chinook('''
SELECT CONCAT(c."FirstName", ' ', c."LastName") AS "Cliente",
      SUM(il."UnitPrice" * il."Quantity") AS "Gasto_Total"
FROM "Customer" AS c
INNER JOIN "Invoice" AS i
ON c."CustomerId" = i."CustomerId"
INNER JOIN "InvoiceLine" AS il
ON i."InvoiceId" = il."InvoiceId"
GROUP BY "Cliente"
ORDER BY "Gasto_Total" DESC
LIMIT 3
''')
clientes_mas_gasto
```

3. Utiliza una CTE para listar todos los artistas que tienen más de 5 álbumes registrados en la base de datos.

```python
artistas_5_albumes = sql.conexion_chinook('''
WITH ArtistaAlbumCinco AS (
    SELECT a."ArtistId", 
           a."Name", 
           COUNT(al."AlbumId") AS "Num_Albumes"
    FROM "Artist" AS a
    LEFT JOIN "Album" AS al
    ON a."ArtistId" = al."ArtistId"
    GROUP BY a."ArtistId", a."Name")
    
SELECT "Name" AS "Artista"
FROM ArtistaAlbumCinco
WHERE "Num_Albumes" > 5
''')
```

```python
artistas_5_albumes
```

4. Encuentra los nombre de las 5 canciones más vendidas junto con el total de ventas de cada una.

```python
canciones_mas_vendidas = sql.conexion_chinook('''
SELECT t."Name" AS "Título",
       ROUND(SUM(il."Quantity"),0) AS "Num_Ventas"
FROM "Track" AS t
INNER JOIN "InvoiceLine" AS il
ON t."TrackId" = il."TrackId"
GROUP BY "Título"
ORDER BY "Num_Ventas" DESC
LIMIT 5
''')
```

```python
canciones_mas_vendidas
```
