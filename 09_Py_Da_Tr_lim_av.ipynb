```python
import pandas as pd
pd.set_option('display.max_columns', None)#Muestre todas las columnas
import numpy as np
```

```python
#DF limpio de clase 07
df = pd.read_csv('Data/online_sales_limpio.csv', parse_dates=['Invoice_Date'])
#Le digo que la columna Invoice_Date es de tipo fecha
```

```python
#10- #Caulclar cuantos días han pasadesde la fecha de cada compra hasta hoy
df_calculos['Resta_Dias'] = df['Invoice_Date'] - pd.Timedelta(days = 5)
```

```python
df_calculos['Suma_Horas'] = df['Invoice_Date'] + pd.Timedelta(hours = 3)
```

```python
#Cuantos días han pasadesde la fecha de cada compra hasta hoy
df_calculos['Tiempo_última_compra'] = (pd.to_datetime('today') - df['Invoice_Date'])
#Cuantas horas han pasado desde la fecha de cada compra hasta hoy
df_calculos.head()
```

```python
#Expresado solo en días quitando las horas, minutos y segundos
df_calculos['Tiempo_última_compra'] = df_calculos['Tiempo_última_compra'].dt.days	
# o directamente df_calculos['Tiempo_última_compra'] = (pd.to_datetime('today') - df['Invoice_Date']).dt.days
df_calculos.head()
```

```python
#11- Separar la columna Invoice_Date en dos columnas, una con la fecha y hora por el espacio

df['Invoice_Date_str'] = df['Invoice_Date'].apply(lambda x: str(x)) #Convierte la fecha a string para mostrarla en el DataFrame y separa fecha y hora
```

```python
df['Invoice_Date_str'] = df['Invoice_Date'].astype(str) #Otra forma de convertir la fecha a string pero más eficiente
```

```python
df.sample(5)
```

```python
#Separo fecha y hora por el espacio

#df['Date'] = df['Invoice_Date_str'].apply(lambda x: x.split(' ')[0])
#df['Hour'] = df['Invoice_Date_str'].apply(lambda x: x.split(' ')[1])
#Unificamos las dos líneas anteriores en una sola
df[['Date', 'Hour']] = df['Invoice_Date_str'].str.split(' ', expand=True)
df.sample(5)
```

```python
#12- Corregir errores de escritura en paypall
df['Payment_Method'].unique()
```

```python
df['Payment_Method'] = df['Payment_Method'].str.replace('paypall', 'paypal')
df['Payment_Method'].unique()
```

```python
#13- Elimino valores negativos de la columna quantity
df['Quantity'] = df['Quantity'].apply(lambda x: np.nan if x < 0 else x)
```

```python
df['Discount'] = df['Discount'].apply(lambda x: np.nan if x > 1 else x)
```

```python
#creamos un diccionario para los meses del año
months = {  
  1: 'January',
  2 : 'February',
  3 : 'March',
  4 : 'April',
  5 : 'May',
  6 : 'June',
  7 : 'July',
  8 : 'August',
  9 : 'September',
  10: 'October',
  11: 'November',
  12: 'December'
}
```

```python
#15- Reemplazar los números de los meses por su nombre en la columna Month
df['Month'] = df['Month'].map(months)
```

```python
#16- Creamos una categortia para shipping_costs en 3 secciones de acuerdo a su valor:
def categoria_shipping(x):
    if x <= 13:
        return 'low'
    elif 13 < x <= 21:
        return 'medium'
    elif x > 21:
        return 'high'
    else:
        return x


  
```

```python
#genero nuea columna Shipping_Cost_Category aplicansdo la funcion creada
df['Shipping_Cost_Category'] = df['Shipping_Cost'].apply(categoria_shipping)
```

```python
#Lo mismo pero optimizado con cortar cut

corte = [5, 13, 21, 30]
categorias = ['low', 'medium', 'high']
df['Shipping_Cost_Category'] = pd.cut(df['Shipping_Cost'], bins=corte, labels=categorias, include_lowest=True)
```

```python
#en una sola línea más optimizada
df['Shipping_Cost_Category'] = pd.cut(df['Shipping_Cost'], bins=[5, 13, 21, 30], labels=['low', 'medium', 'high'], include_lowest=True)
```
