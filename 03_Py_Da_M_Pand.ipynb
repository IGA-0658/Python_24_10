## Loc e iloc

-- Permite acceder a datos concretos del DF

* LOC -> accder a los datos a traves de etiquetas

```python

df.loc[etiquetas_filas,etiquetas_columnas]

```


* iloc -> accder a los elementos a traves de su indice

```python

df.iloc[indice_filas,indice_columnas]

```

Para no repertir formula, crea una variable con el calculo.

### assign()

Genero mÃ¡s de una columna a la vez, tantas columnas como quiera, siempre al final.

df.assign(nueva_col_1 = expresion_1, nueva_col_2 = expresion_2)



````python
copia.assign(DescuentoAplicadoAssign_1 = calculo_descuento, DescuentoAplicadoAssign_2 = calculo_descuento)
````

````python
#Assing NO sobre escribe, Si quiero sobrescribir copia con los cambios guardo como nuevo DF

copia = copia.assign(DescuentoAplicadoAssign_1 = calculo_descuento, DescuentoAplicadoAssign_2 = calculo_descuento)
````

13)->Crear copia del DF, elimina columnas StockCode y CustomerID, y ordena el resultado por Country y Quantity.

````python
#Copia
no_stock_customer = df.copy()
#Elimino
no_stock_customer.drop(['StockCode', 'CustomerID'], axis=1, inplace=True)
#Ordeno
no_stock_customer.sort_values(by=['Country', 'Quantity'])
````
