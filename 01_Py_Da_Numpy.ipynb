```python
#Por ejemplo un array "ventas" con las ventas de 12 meses
ventas = np.random.randint(0,1000, (5,12))
print(ventas)
print("----------")
```

###Analisis estadístico descriptivo

```python
media_ventas = np.mean(ventas)
media_ventas
```

```python
#media ventas 1º comercial
media_ventas_comercial1 = np.mean(ventas[0])
media_ventas_comercial1
```

```python
media_ventas_totales = np.median(ventas)
media_ventas_totales
```

```python
minimo = np.min(ventas)
minimo
```

```python
desviacion_std = np.std(ventas)
desviacion_std
```

```python
ventas_superiores_media = np.where(ventas > media_ventas)
ventas_superiores_media
#muestra las posiciones
```
